```python
import pandas as pd
from mlflow.tracking import MlflowClient
import matplotlib.pyplot as plt
client = MlflowClient()
```

```python
hashed = "5515c927593747b585616cad56d0efba"
local_path = client.download_artifacts(hashed, "metrics.csv", f"artifacts/{hashed}/")
cols = ["Gradient Penalty",	"Wasserstein Distance Estimate", "Iteration number", "GPU Memory", "EOF Coefficient L2", "Divergence L2", "Vorticity L2", "Content L2", "Generator loss"]

path = f"artifacts/{hashed}/artifacts/metrics.csv"
df_eof = pd.read_csv(f"artifacts/{hashed}/metrics.csv", usecols = cols)
```

```python
hashed = "89ab7d6b0c9e447fa7ae3ca30044f06c"
local_path = client.download_artifacts(hashed, "metrics.csv", f"artifacts/{hashed}/")
cols = ["Gradient Penalty",	"Wasserstein Distance Estimate", "Iteration number", "GPU Memory", "EOF Coefficient L2", "Divergence L2", "Vorticity L2", "Content L2", "Generator loss"]

path = f"artifacts/{hashed}/artifacts/metrics.csv"
df_content_g16_c8 = pd.read_csv(f"artifacts/{hashed}/metrics.csv", usecols = cols)
```

```python
fig, ax = plt.subplots(2, 3, figsize=(20, 10))
labels = ["Extended", "EOF+Content", "Content", "Content G16 C8"]
ax[0, 0].plot(df_eof["Iteration number"], df_eof["Wasserstein Distance Estimate"].rolling(window=10).mean(), label=labels[0])
ax[0, 0].plot(df_extended["Iteration number"], df_extended["Wasserstein Distance Estimate"].rolling(window=10).mean(), label=labels[1])
ax[0, 0].plot(df_content["Iteration number"], df_content["Wasserstein Distance Estimate"].rolling(window=10).mean(), label=labels[2])
ax[0, 0].plot(df_content_g16_c8["Iteration number"], df_content_g16_c8["Wasserstein Distance Estimate"].rolling(window=10).mean(), label=labels[3])

ax[0, 0].semilogy()
ax[0, 0].set_title("Wasserstein Distance")
ax[0, 0].set_xlabel("Iteration Number")
ax[0, 0].legend()

ax[1, 0].plot(df_eof["Iteration number"], df_eof["EOF Coefficient L2"].rolling(window=10).mean(), label=labels[0])
ax[1, 0].plot(df_extended["Iteration number"], df_extended["EOF Coefficient L2"].rolling(window=10).mean(), label=labels[1])
ax[1, 0].plot(df_content["Iteration number"], df_content["EOF Coefficient L2"].rolling(window=10).mean(), label=labels[2])
ax[1, 0].plot(df_content_g16_c8["Iteration number"], df_content_g16_c8["EOF Coefficient L2"].rolling(window=10).mean(), label=labels[3])

ax[1, 0].semilogy()
ax[1, 0].set_title("EOF Coefficient L2")
ax[1, 0].set_xlabel("Iteration Number")
ax[1, 0].legend()

ax[0, 1].plot(df_eof["Iteration number"], df_eof["Divergence L2"].rolling(window=10).mean(), label=labels[0])
ax[0, 1].plot(df_extended["Iteration number"], df_extended["Divergence L2"].rolling(window=10).mean(), label=labels[1])
ax[0, 1].plot(df_content["Iteration number"], df_content["Divergence L2"].rolling(window=10).mean(), label=labels[2])
ax[0, 1].plot(df_content_g16_c8["Iteration number"], df_content_g16_c8["Divergence L2"].rolling(window=10).mean(), label=labels[3])

ax[0, 1].semilogy()
ax[0, 1].set_title("Divergence L2")
ax[0, 1].set_xlabel("Iteration Number")
ax[0, 1].legend()

ax[1, 1].plot(df_eof["Iteration number"], df_eof["Vorticity L2"].rolling(window=10).mean(), label=labels[0])
ax[1, 1].plot(df_extended["Iteration number"], df_extended["Vorticity L2"].rolling(window=10).mean(), label=labels[1])
ax[1, 1].plot(df_content["Iteration number"], df_content["Vorticity L2"].rolling(window=10).mean(), label=labels[2])
ax[1, 1].plot(df_content_g16_c8["Iteration number"], df_content_g16_c8["Vorticity L2"].rolling(window=10).mean(), label=labels[3])

ax[1, 1].semilogy()
ax[1, 1].set_title("Vorticity L2")
ax[1, 1].set_xlabel("Iteration Number")
ax[1, 1].legend()

ax[0, 2].plot(df_eof["Iteration number"], df_eof["Content L2"].rolling(window=10).mean(), label=labels[0])
ax[0, 2].plot(df_extended["Iteration number"], df_extended["Content L2"].rolling(window=10).mean(), label=labels[1])
ax[0, 2].plot(df_content["Iteration number"], df_content["Content L2"].rolling(window=10).mean(), label=labels[2])
ax[0, 2].plot(df_content_g16_c8["Iteration number"], df_content_g16_c8["Content L2"].rolling(window=10).mean(), label=labels[3])

ax[0, 2].semilogy()
ax[0, 2].set_title("Content L2")
ax[0, 2].set_xlabel("Iteration Number")
ax[0, 2].legend()

ax[1, 2].plot(df_eof["Iteration number"], df_eof["Gradient Penalty"].rolling(window=10).mean(), label=labels[0])
ax[1, 2].plot(df_extended["Iteration number"], df_extended["Gradient Penalty"].rolling(window=10).mean(), label=labels[1])
ax[1, 2].plot(df_content["Iteration number"], df_content["Gradient Penalty"].rolling(window=10).mean(), label=labels[2])
ax[1, 2].plot(df_content_g16_c8["Iteration number"], df_content_g16_c8["Gradient Penalty"].rolling(window=10).mean(), label=labels[3])

ax[1, 2].semilogy()
ax[1, 2].set_title("Gradient Penalty")
ax[1, 2].set_xlabel("Iteration Number")
ax[1, 2].legend()

plt.suptitle("Losses", fontsize=20)
plt.tight_layout()
plt.savefig("artifacts/Losses_Initial_Experiments")
```
